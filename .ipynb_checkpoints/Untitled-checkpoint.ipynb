```python
import numpy as np
import matplotlib.pyplot as plt

# Load the NPZ file
data = np.load('CWRU_48k_load_1_CNN_data.npz')

# See what's inside
print("Keys in the file:")
print(data.files)

# Load a specific dataset (common names: 'X_train', 'y_train', 'X_test', 'y_test')
# Let's explore the structure
for key in data.files:
    arr = data[key]
    print(f"\n{key}:")
    print(f"  Shape: {arr.shape}")
    print(f"  Data type: {arr.dtype}")
    print(f"  Sample values: {arr.flatten()[:10]}")
```
